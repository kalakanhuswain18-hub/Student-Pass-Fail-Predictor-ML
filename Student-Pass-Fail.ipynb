```python
from sklearn.linear_model import LogisticRegression 
import pandas as pd 
```

```python
x=[[1],[3],[5],[7],[9]]
y=[0,0,1,1,1]
model=LogisticRegression()
model.fit(x,y)
hours=float(input("Enter Study hours : "))
# new_data=model.DataFrame([[hours ,sleep]] , columns=["Hours" ,"Sleep"])
result=model.predict([[hours]])[0]
if result == 1 :
    print(f"based on {hours} hours , you are likely pass in exam !!GOOD!!🎉")
else :
    print(f" based on {hours} hours , you are likely fail in exam !!Try Again!!😞")
```

```python
data=pd.read_csv("student_pass_fail.csv")
data
```

```python
X=data[["Hours" ,"Revision"]]
Y=data["Pass"]
model=LogisticRegression()
model.fit(X,Y)
hours=float(input("Enter Study Hours : "))
revise=float(input("Enter Revise Hours : "))
new_data=pd.DataFrame([[hours , revise]] , columns=["Hours" , "Revision"])
pred=model.predict(new_data)[0]
if pred == 1:
    print(f"Congrats 🎉 Based on {hours} hours of study and {revise} hours of revision ,  you are pass in exam !!GOOD!! ")
else :
    print(f"Ohh No😞!! Based on {hours} hours of study and {revise} hours of revision , you  are failed in exam !!BAD!! Try again")
prob = model.predict_proba(new_data)[0][1]
print(f"Pass Probability: {prob:.2f}")    
```
